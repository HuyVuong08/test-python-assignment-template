Given two integer arrays array1 and array2, return an array of the common integers in two arrays. Each element in the result must be unique and you may return the result in any order. Make sure your code can pass the test on LeetCode to get the full mark.

Example :

Input: `array1 = [3,9,5,7]`, `array2 = [2,5,9]`

Output: `[5, 9]` or `[9, 5]`


```python
def intersection(array1, array2):
    """
    Find common integers between two arrays.
    
    Args:
        array1: First integer array
        array2: Second integer array
    
    Returns:
        List of unique common integers
    """
    return list(set(array1) & set(array2))


# Test with the example
array1 = [3, 9, 5, 7]
array2 = [2, 5, 9]
result = intersection(array1, array2)
print(result)
```
